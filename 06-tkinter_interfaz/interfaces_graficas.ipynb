Hacer lo mismo, pero con resta, multiplicación y división.

```python
# Base de trabajo

from tkinter import *

# función de sumar 
def sumar():
    r.set(float(n1.get()) + (float(n2.get())))

# función de restar 
def restar():
    r.set(float(n1.get()) - (float(n2.get())))

# función de multiplicar 
def multiplicar():
    r.set(float(n1.get()) * (float(n2.get())))

# función de dividir 
def dividir():
    if float(n2.get()) == 0:
        r.set('¡Error, no se puede dividir por 0!')
    else:
        r.set(float(n1.get()) / (float(n2.get())))

# Configuración de la raiz o ventana principal (instancia de la clase Tk)

root = Tk()

# variables de texto 
n1 = StringVar()
n2 = StringVar()
r = StringVar()

# creamos los campos de texto 
Label(root, text = 'Número 1: ').pack()
Entry(root, justify= 'center', textvariable= n1).pack()

Label(root, text = 'Número 2: ').pack()
Entry(root, justify= 'center', textvariable= n2).pack()

Label(root, text = 'Resultado: ').pack()
Entry(root, justify= 'center', textvariable= r, state= 'disabled', width = 30).pack() #con disabled no puedes meter resultado

# botón de sumar 
Button(root, text = 'Sumar', command = sumar).pack(side = 'left') # Puedes mover el botón y apilarlo con pack(side = 'left')

# botón de restar
Button(root, text = 'Restar', command = restar).pack(side = 'left')

# botón de multiplicar 
Button(root, text = 'Multiplicar', command = multiplicar).pack(side = 'left')

# botón de dividir
Button(root, text = 'Dividir', command = dividir).pack(side = 'left')

# Bucle de la aplicación para que se ejecute todo el rato (se crea una ventana, para acabar la cierras)

root.mainloop()
```

Widget Radiobutton - Botones de selección única

```python
# Base de trabajo

from tkinter import *

# función para mostrar el valor de los radiobuton en el monitor

def seleccionar():
    monitor.config(text= f'{opcion.get()}')

# función para resetear los valores

def reset():
    opcion.set(None)
    monitor.config(text= '')

# Configuración de la raiz o ventana principal (instancia de la clase Tk)

root = Tk()

# variable para el valor de cada uno de los botones

opcion = IntVar() # variable entera, habiamos visto stringbar. hay boolean tambien etc

# configuración de los radiobutton

Radiobutton(root, text = 'Opción 1', variable = opcion, value = 1, command = seleccionar).pack() # Si sólo pones 1 se marca automáticamente.
Radiobutton(root, text = 'Opción 2', variable = opcion, value = 2, command = seleccionar).pack() 
Radiobutton(root, text = 'Opción 3', variable = opcion, value = 3, command = seleccionar).pack() 

# monitor para ver los valores

monitor = Label(root)
monitor.pack()

# Botón de reset 
Button(root, text = 'reset', command = reset).pack()

# Bucle de la aplicación para que se ejecute todo el rato (se crea una ventana, para acabar la cierras)

root.mainloop()
```

Widget Menu (Menús)

```python
# Base de trabajo
from tkinter import *

# Configuración de la raiz o ventana principal (instancia de la clase Tk)
root = Tk()

# configuración de la barra de menú principal (menubar)
menubar = Menu(root)
root.config(menu = menubar)

#creamos los elementos del menú
#creamos los submenús. creamos elementos en el menú (submenús) con add_command
filemenu = Menu(menubar, tearoff= 0) #tearoff para quitar la lágrima al pulsar "- -"
filemenu.add_command(label='Nuevo')
filemenu.add_command(label='Abrir')
filemenu.add_command(label='Guardar')
filemenu.add_separator() # Para separar los submenús
filemenu.add_command(label='Salir', command= root.destroy) # podemos poner root.quit en vez del destroy, pero dejaría la ventana abierta y acaba petando

editmenu = Menu(menubar, tearoff= 0)
editmenu.add_command(label= 'Cortar')
editmenu.add_command(label= 'Copiar')
editmenu.add_command(label= 'Pegar')


helpmenu = Menu(menubar, tearoff= 0)
helpmenu.add_command(label= 'Ayuda')
helpmenu.add_separator()
helpmenu.add_command(label= 'Acerca de...')


#añadimos los elementos al menú
menubar.add_cascade(label= 'Archivo', menu= filemenu)
menubar.add_cascade(label= 'Editar', menu= filemenu)
menubar.add_cascade(label= 'Ayuda', menu= filemenu)


# Bucle de la aplicación para que se ejecute todo el rato (se crea una ventana, para acabar la cierras)
root.mainloop()
```

AskOpenFile: abrir archivo desde el sitio donde trabajamos

```python
# Base de trabajo

from tkinter import *
from tkinter import filedialog as FD
from pathlib import Path

# función para obtener la ruta de un archivo
def test():
    archivo = FD.askopenfile(title= 'Abrir archivo', initialdir= '.',
                              filetypes= (('Archivos de texto', '*.txt'),
                                          ('Archivo enriquecido', '*.doc'))) # titulo ventana, directorio inicial donde buscar (si pones '.' es donde estás ahora) y tipo archivos en formato tupla acabar en , la tupla siempre
    # para todos tipos de archivos "*.*"
    # print(archivo) # al darle click a un arvhivo te printa la ruta absoluta
    if archivo: # Si hay algo en el archivo va a imprimir el contenido de este
        file_path = Path(archivo)
        contenido = file_path.read_text(encoding= 'utf-8')
        print(contenido)


# Configuración de la raiz o ventana principal (instancia de la clase Tk)

root = Tk()


#generador de eventos (Button)
Button(root, text = 'Haz click!', command= test).pack()

# Bucle de la aplicación para que se ejecute todo el rato (se crea una ventana, para acabar la cierras)

root.mainloop()
```
