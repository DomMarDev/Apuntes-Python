# Configuracin de la raz

Vamos a comenzar por lo ms esencial, la raz o base de la interfaz grfica.

Recordad que la raz es el contenedor base de todos los widgets que forman la interfaz, no tiene tamao propio sino que se adapta a los widgets que contiene:

```python
# importamos del mdulo tkinter todos los witches (1)
from tkinter import *

# creamos una raz a partir del mdulo Tk(1)
root = Tk()

# configuracin del ttulo(2)
root.title('Hola Mundo!!')

# configuracin de un icono(2)
root.iconbitmap('img/hola.ico')
# linux => root.iconbitmap('@hola.bm')

# configuracin de aspectos de la dimensin de la ventana (resize)(2)
root.resizable(True, True)
# valores: ancho, alto  / True, False

# ponemos en marcha el programa principal con el mtodo mainloop()(1)
root.mainloop()
```

# Widget frame (marco)

Los Frames son marcos contenedores de otros widgets. Pueden tener tamao propio y posicionarse en distintos lugares de otro contenedor (ya sea la raz u otro marco):

```python
from tkinter import *

# Configuracin de la raz (1)
root = Tk()
root.title("Hola mundo")
root.resizable(1,1)
root.iconbitmap('img/hola.ico')

# configuracin del marco, le pasamos el parmetro root y las medidas
frame = Frame(root,width=480, height=320)
# frame.config(width=480, height=320)

# le decimos al frame que se empaquete dentro del root, se distribuya
# frame.pack()
# frame.pack(side='right', anchor='center')
# frame.pack(fill='x')
# frame.pack(fill='y', expand=1)
frame.pack(fill='both', expand=1)

frame.config(cursor="pirate")

frame.config(bg="lightblue")

frame.config(bd=25)

frame.config(relief="sunken")


root.config(cursor="arrow")
root.config(bg="blue")
root.config(bd=15)
root.config(relief="ridge")


# Finalmente bucle de la apliacin (1)
root.mainloop()
```

- Valores para *side*: right, left, top, bottom
- Valores para *anchor*: n, s, e, w, ne, nw, se, sw, center
- Valores de relleno (*fill*): x, y, both
- Parmetro de expansin (*expand*): 1 para que se expanda junto con el root
- Otros parmetros de configuracin: cursor, bg(background), bd(border en px), relief(tipo de border)

- Mejor forma de colocar los elementos de un formulario

```python
from tkinter import *

# configuracin de la raz
root = Tk()

# creamos el primer marco con su entry y su label
frame1 = Frame(root)
frame1.pack()

entry1 = Entry(frame1)
entry1.pack(side='right')

label1 = Label(frame1, text = 'Nombre  ')
label1.pack(side = 'left')
# creamos el segundo marco con su entry y su label
frame2 = Frame(root)
frame2.pack()

entry2 = Entry(frame2)
entry2.pack(side='right')

label2 = Label(frame2, text = 'Apellidos')
label2.pack(side = 'left')



# bucle de la aplicacin
root.mainloop()
```

- Mejorando el posicionamiento usando cuadrcula (grid())
- *sticky* alinea el campo (n,s,w,e)
- *padx* y *pady* genera un padding (relleno) en los ejes x e y (px)

```python
from tkinter import *

# configuracin de la raz
root = Tk()


entry1 = Entry(root)
entry1.grid(row=0, column=1)

# podemos justificar el campo de entrada
entry1.config(justify='right')
# podemos desactivar el campo de entrada
# entry1.config(state='disabled')
# podemos poner un campo de entrada de contrasea
entry1.config(show='*')

label1 = Label(root, text = 'Nombre muy muy largo  ')
label1.grid(row=0, column=0, sticky = 'w', padx=5,pady = 5)



entry2 = Entry(root)
entry2.grid(row=1, column=1)

label2 = Label(root, text = 'Apellidos')
label2.grid(row=1, column=0, sticky = 'w', padx=5, pady= 5)


# bucle de la aplicacin
root.mainloop()
```

# Widget Checkbutton (Seleccionable)
Con los radiobutton vimos que el usuario puede marcar una opcin de entre varias, pero si queremos simplemente proponer una nica opcin es mejor utilizar un botn de seleccin. Son bastante parecidos. Vamos a crear un par de checkbuttons para por ejemplo, pedir a un usuario como quiere que le sirvan un caf, l tendr que marcar si quiere leche, y tambin si quiere azcar:

```python
# Base de trabajo

from tkinter import *

#  configuracin de la raz
root = Tk()
root.title('Cafetera')
root.config(bd=15)

# configuracin de los botones check
leche = IntVar() # 1 s, 0 no
azucar = IntVar()
# leche = BooleanVar() # 1 s, 0 no
# azucar = BooleanVar()

# imagen de la cafetera
imagen = PhotoImage(file = 'img/imagen.gif')
Label(root, image = imagen).pack(side='left')

frame = Frame(root)
frame.pack(side='right')

Label(frame, text = 'Cmo quieres el caf?: ').pack(anchor='w')
Checkbutton(frame, text = 'Con leche', variable = leche, onvalue = 1, offvalue =0).pack(anchor='w')
Checkbutton(frame, text = 'Con azcar', variable = azucar, onvalue = 1, offvalue = 0).pack(anchor='w')


# # bucle de la aplicacin
root.mainloop()
```

```python
# Base de trabajo

from tkinter import *

# funciones
def seleccionar():
    cadena = ''
    if (leche.get()):
        cadena += 'Con leche' 
    else:
        cadena += 'Sin leche'
    
    if (azucar.get()):
        cadena += ' y con azcar'
    else:
        cadena += ' y sin azcar'
        
    monitor.config(text = cadena)
#  configuracin de la raz
root = Tk()
root.title('Cafetera')
root.config(bd=15)

# configuracin de los botones check
leche = BooleanVar() # 1 s, 0 no
azucar = BooleanVar()

# imagen de la cafetera
imagen = PhotoImage(file = 'img/imagen.gif')
Label(root, image = imagen).pack(side='left')

frame = Frame(root)
frame.pack(side='right')

Label(frame, text = 'Cmo quieres el caf?: ').pack(anchor='w')
Checkbutton(frame, text = 'Con leche', variable = leche, onvalue = True, offvalue = False, command = seleccionar).pack(anchor='w')
Checkbutton(frame, text = 'Con azcar', variable = azucar, onvalue = True, offvalue = False, command = seleccionar).pack(anchor='w')


# monitor
monitor = Label(frame)
monitor.pack()

# # bucle de la aplicacin
root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog as FileDialog
from pathlib import Path

def test():
    fichero = FileDialog.askopenfilename(
        title="Abrir un fichero", 
        initialdir='C:', 
        filetypes=(('Ficheros de texto', '*.txt'), ('Ficheros de texto avanzado', '*.word'))
    )

    if fichero:  # Verificar si el usuario seleccion un archivo
        file_path = Path(fichero)
        # Leer el contenido del archivo
        contenido = file_path.read_text()
        print(contenido)

root = Tk()

Button(root, text="Clcame", command=test).pack()

root.mainloop()
```

## AskSaveAsFile

```python
from tkinter import *
from tkinter import filedialog as FileDialog
from pathlib import Path

ruta = "" # La utilizaremos para almacenar la ruta del fichero

def nuevo():
    global ruta
    mensaje.set("Nuevo fichero")
    ruta = ""
    texto.delete(1.0, "end") # borra el primer carcter hasta el ltimo
    root.title("Mi editor") # reiniciamos el nombre del ttulo

def abrir():
    global ruta
    mensaje.set("Abrir fichero")
    ruta = FileDialog.askopenfilename(
        initialdir='.', 
        filetypes=(("Ficheros de texto", "*.txt"),),
        title="Abrir un fichero de texto")

    if ruta != "":
        ruta_path = Path(ruta)
        contenido = ruta_path.read_text()
        texto.delete(1.0, 'end')
        texto.insert('insert', contenido) # modo para insertar texto
        root.title(ruta + " - Mi editor") # cambiamos el ttulo de la ventana

def guardar():
    mensaje.set("Guardar fichero")
    if ruta != "": # si ya tenemos la ruta, o sea, el archivo ya existe
        contenido = texto.get(1.0, 'end-1c') # desde el carcter 1 hasta el final menos 1 carcter
        ruta_path = Path(ruta)
        ruta_path.write_text(contenido)
        mensaje.set("Fichero guardado correctamente")
    else:
        guardar_como()

def guardar_como():
    global ruta
    mensaje.set("Guardar fichero como")

    fichero = FileDialog.asksaveasfile(title="Guardar fichero", 
        mode="w", defaultextension=".txt")

    if fichero is not None:
        ruta = fichero.name
        contenido = texto.get(1.0, 'end-1c')
        ruta_path = Path(ruta)
        ruta_path.write_text(contenido)
        mensaje.set("Fichero guardado correctamente")
    else:
        mensaje.set("Guardado cancelado")
        ruta = ""

# Configuracin de la raz
root = Tk()
root.title("Mi editor")

# Men superior
menubar = Menu(root)
filemenu = Menu(menubar, tearoff=0)
filemenu.add_command(label="Nuevo", command=nuevo)
filemenu.add_command(label="Abrir", command=abrir)
filemenu.add_command(label="Guardar", command=guardar)
filemenu.add_command(label="Guardar como", command=guardar_como)
filemenu.add_separator()
filemenu.add_command(label="Salir", command=root.destroy)
menubar.add_cascade(menu=filemenu, label="Archivo")

# Caja de texto central
texto = Text(root)
texto.pack(fill="both", expand=1)
texto.config(bd=0, padx=6, pady=4, font=("Consolas",12))

# Monitor inferior
mensaje = StringVar()
mensaje.set("Bienvenido a tu Editor")
monitor = Label(root, textvar=mensaje, justify='left')
monitor.pack(side="left")

root.config(menu=menubar)
# Finalmente bucle de la aplicacin
root.mainloop()
```
