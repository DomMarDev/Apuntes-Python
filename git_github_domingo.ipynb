Git y GitHub son dos herramientas que nos permiten controlar las distintas versiones cuando nos dedicamos al desarrollo de software. Son Gestores de Control de Versiones.


* Git: es un sistema de control de versiones que nos permite rastrear los cambios que hemos hecho en un conjunto de archivos.
* GitHub: Servicio de hosting que nos permite almacenar proyectos de desarrollo de software y control de versiones usando Git.

# Conceptos Básicos

## Sistemas de Control de versiones

Sistema encargado de administrar los cambios realizados en programas o conjuntos de archivos.

## Repositorio

Colección de archivos de distintas versiones de un proyecto. Podemos tener repositorios locales o repositorios remotos.

## Commit

Componentes básicos de la línea del tiempo de un proyecto Git. Son como un registro o "foto" del estado de un proyecto en un momento determinado.



## Git Bash

Herramienta que permite ejecutar comandos de Git. Es una consola de línea de comandos.

### Moverse dentro de la terminal:

## Configurar nuestro usuario y e-mail de Git

Para registrar nuestro usuario y e-mail en Git y así poder controlar quién realiza cambios o "commits", usaremos los siguientes comandos:

* Nombre de usuario: `git config --global user.name "[nombre del usuario]`".
* E-mail del usuario: `git config --global user.email "[e-mail]"`.

En el e-mail del usuario, cuando tengamos la cuenta de GitHub usaremos el mail privado que nos da GitHub. Para saber el usuario o el mail que tenemos configurados usaremos:

* `git config user.name` o `git config user.email`

## Áreas de trabajo y estados de los archivos


### Directorio de trabajo (Modificada o Modified)

La carpeta del proyecto que contiene los archivos y el directorio .git del repositorio. La versión del archivo contiene cambios que nos parte del repositorio y no se ha añadido al área de preparación.

Ejemplo:
Abrimos con visual la carpea prueba1 -> hola.py
 si ves una U es que no lo está ni guardando (untracked).

### Área de preparación (Preparada o Staged)

Área intermedia entre el Directorio de Trabajo y el Repositorio. Es el conjunto de archivos y cambios que serán incluidos en el próximo commit. Si la versión del archvio contiene cambios que nos parte del repositorio pero se ha añadido al área de preparación.


### Repositorio (Comprometida o Committed)

Directorio que contiene los metadatos y las versiones de nuestro proyecto. El directorio de git (.git) es la parte del repositorio que se copia cuando clonamos un repositorio en nuestro ordenador. Es la parte más importante de Git. Si la versión del archivo ya se encuentra en el directorio Git ya se encuentra Confirmada.



### Comandos para verificar el estado

Para mirar el estado del repositorio usaremos el comando:

* `git status`

Para añadir un archivo al área de preparación:

* `git add [nombre del archivo}` o bien `git add .` para añadir todo.

Ahora ya están preparados los archivos, tienes una A al lado del archivo en visual. Peendiente de commit.


Para revertir el estado del archivo del área del preparación

* `git rm --cached [nombre del archivo]`

Para realizar el commit y añadir los cambios del repositorio

* `git commit -m "Texto descriptivo del archivo o sus cambios"`

o bien

* `git commit` y VSCode se encargará de realizar el resumen del archivo

(Aclaración)te lleva al archivo para configurarlo, en el commit:editmsg poner el texto del commit, guardas y cierras ese archivo.


Cuando realizamos un commit vemos una estructura como ésta:



La línea roja nos indica su Hash o identificador único SHA (Secure Hash Algorithm). Identifica los cambios realizados, dónde se realizaron y quién los realizó.
(Aclaración) git log aparece un hash en cabecera (identificador), un autor, mail, fecha creación, texto sobre qué se ha hecho (commit)


Para visualizar el log de commits, usaremos:

* `git log` y si queremos usar un operador para verlo en modo resumen `git log --oneline`.
* También usar `git log -p` para ir mostrando líneas una a una.

### Asociar un editor de código diferente

Para cambiar el editor de codigo (si tienes otro) y poner en la configuración el visual studio (si no se te abre el archivo msg)

* `git config --globarl core.editor "code --wait"`.

### Cambiar o editar el texto de un commit

Para cambiar el texto de un commit:

* `git commit --amend`, recomendable sólo hacer el repositorio local, para evitar problemas en el repositorio remoto.

Para revertir el último commit añadido: Las reversiones siempre sobre la ultima versión!

* `git reset --soft HEAD~1` 
* (Con `soft` mantiene los cambios del archivo y lo deja a la espera de guardar. `--hard` elimina todos los cambios).

Sobretodo, hacerlo sobre el último commit, no sobre versiones históricas anteriores.

Después agregaremos los cambios con `git add .` y volveremos a realizar el commit con `git commit`.


## Ramas en Git

Definición: Las ramas son líneas independientes de desarrollo en el repositorio.

M (principal):
    - Rama 1 - se hacen X commits
    - Rama 2 - "
    - Rama 3 - "
Luego las ramas se pueden incorporar al Master y puedes eliminar las ramas una vez se han incorporado en el principal.

orden: git add . > git commit > modificas sentencia > git status y git log


### Crear ramas a partir de la rama principal

* `git branch [nombre de la rama]` crear ramas.
* `git branch` para visualizar las ramas que tenemos (en forma de lista).


### Cambiar de ramas

* `git checkout [nombre de la rama]`

### Creando una rama y cambiando al mismo tiempo

Desde la rama `main`, entraremos el comando:

* `git checkout -b [nombre de la rama nueva]`

### Cambiando el nombre de una rama

Desde la rama que queremos cambiar el nombre:

* `git branch -m [nuevo nombre]`

Desde otra rama diferente:

* `git branch -m [nombre antiguo] [nombre nuevo]`

### Eliminando ramas

Desde fuera de la rama a eliminar:

* `git branch -d [nombre de la rama]`

### Fusionar ramas

Para combinar los cambios realizados en la rama con la rama principal:

* `git merge [nombre de la rama a fusionar]`, estando en la rama que va a recibir la fusión.

Si encontramos problemas o conflictos podemos dejar a VSCode que los solucione, eligiendo qué fusión va a suceder, y después usaremos el comando `git merge --continue`.

Fusionar rama con la principal:
    1 ir a la rama que va a recibir la fusión (la main)
    2. git merge (nombre de la rama que se va a fusionar)
        Si haces git log: 
            - lista de logs de la rama y de la main.

Si aparece un error de combinación darle a Editor de cambios:
    Escoger la combinación que se quieres
    Al final completar la fusión y confirmas.

## Clonar un repositorio remoto en local

Entramos en git bash, dentro de la carpeta que hicimos:

* `git clone [URL del repositorio en GitHub]`

* Con `git log` podremos ver los commits que han sucedido en ese repositorio.

Para saber donde está conectado/ vinculado con ese repositorio:

* Con `git remote` o `git remote -v` visualizaremos si han creado los métodos para buscar cambios y enviar cambios (fetch y push).

(Aclaración) La rama principal de origin -> cuenta de github


## Enviar cambios a GitHub

* `git push origin main`, nos pedirá que hagamos login en nuestro browser o bien configuraremos un Token en GitHub (perfil/Developer settings).


## Actualizar nuestro repositorio local con GitHub

Cambios en remoto, guardamos. Pero en git bash como saber si se ha realizado un cambio, e importarlo en local?

* `git fetch` nos ayuda a verificar si se han realizado cambios en el repositorio remoto.

* `git checkout origin/main` podemos cambiar a la línea del repositorio remoto.

    (Aclaración) para ir a la rama de remoto origen y en visual estarás en remoto.

* `git checkout main` para volver a la rama en local.

* `git pull origin main` con esto sobreescribes el repositorio local con el remoto
