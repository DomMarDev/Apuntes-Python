Calculadora basica unificada:

```python

# Base de trabajo:
from tkinter import  *

# Funcion de sumar:
def sumar():
    r.set(float(n1.get())+ float(n2.get()))

# Funcion de restar:
def restar():
    r.set(float(n1.get()) - float(n2.get()))
    
# Funcion de multiplicar:
def multiplicar():
    r.set(float(n1.get()) * float(n2.get()))

# Funcion de dividir:
def dividir():
    
    if float(n1.get()) <= 0:
        r.set('Error')
    else:
        r.set(float(n1.get()) / float(n2.get()))


# Configuracion de la raiz(ventana principal):
root = Tk()

# Variables de texto:
n1 = StringVar()
n2 = StringVar()
r = StringVar()

# Creamos los campos de buttom:
Label(root, text='Numero1').pack()
Entry(root, justify='center', textvariable=n1).pack()

Label(root, text='Numero2').pack()
Entry(root, justify='center', textvariable=n2).pack()

Label(root, text='Resultado').pack()
Entry(root, justify='center', textvariable=r, state='disabled').pack()

# Botones:
Button(root, text='Sumar', command= sumar).pack(side='left')
Button(root, text='Restar', command= restar).pack(side='left')
Button(root, text='Multiplicar', command= multiplicar).pack(side='left')
Button(root, text='Dividir', command= dividir).pack(side='left')


# Bucle de la aplicacion:
root.mainloop()
```

Widget RadioButton - Botones de seleccion unica:

```python
# Base de trabajo:
from tkinter import  *

# Funcion para mostrar el valor de los radiobutton en el monitor:
def seleccionar():
    monitor.config(text=f'{opcion.get()}')

# Funcion para resetear valores:
def reset():
    opcion.set(None)
    monitor.config(text='')
    
# Configuracion de la raiz(ventana principal):
root = Tk()

# Variable para el valor de cada uno de los botones:
opcion = IntVar()

# Configuracion de los radio button:
Radiobutton(root, text='Opcion 1', variable=opcion, value=1, command=seleccionar).pack()
Radiobutton(root, text='Opcion 2', variable=opcion, value=2, command=seleccionar).pack()
Radiobutton(root, text='Opcion 3', variable=opcion, value=3, command=seleccionar).pack()


# Monitor para ver los valores:
monitor = Label(root)
monitor.pack()

# Botón de reset:
Button(root, text='Reset', command=reset).pack()

# Bucle de la aplicacion:
root.mainloop()
```

Widget Checkbutton - Botones de seleccion multiple:

```python
# Base de trabajo:
from tkinter import  *

# Funcion para configurar el monitor:
def seleccionar():
    cadena = ''
    if (leche.get()):
        cadena += 'Con leche'
    else:
        cadena += 'Sin leche'
    if (azucar.get()):
        cadena += ' y con azucar'
    else:
        cadena += ' y sin azucar'
    
    monitor.config(text=cadena)

# Configuracion de la raiz(ventana principal):
root = Tk()
root.title('Cafeteria')
root.config(bd=15)

# Variables para asociar los checkbutton:
leche = BooleanVar()
azucar = BooleanVar()

# Imagen de la cafeteria:
imagen = PhotoImage('tenor.gif')
Label(root, image=imagen).pack(side='left')

frame = Frame(root)
frame.pack(side='right')

Label(frame, text='¿Como quieres el café?').pack(anchor='w')
Checkbutton(frame, text='Con leche', variable=leche, onvalue=True, offvalue=False, command=seleccionar).pack(anchor='w')
Checkbutton(frame, text='Con azucar', variable=azucar, onvalue=True, offvalue=False, command=seleccionar).pack(anchor='w')

# Monitor:
monitor = Label(frame)
monitor.pack()

# Bucle de la aplicacion:
root.mainloop()
```
