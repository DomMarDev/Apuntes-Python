```python
# Realizamos un scrapping a una tabla de la wikipedia
df = pd.read_html('https://web.archive.org/web/20220717170349/https://en.wikipedia.org/wiki/List_of_countries_by_past_fertility_rate')
```

```python
df # son varias tablas, pero queremos la de la posición 2
```

```python
df[2]
```

```python
# guardamos el dataframe en una variable
fertility_rate = df[2]
fertility_rate.head(5)
```

```python
# Renombrar la cabecera (método de limpieza de datos)
fertility_rate.rename(columns= {'Country/dependent territory': 'Country'}, inplace= True) # cambias una por otra
fertility_rate.head()
```

```python
# Hacemos una consulta
fertility_rate[['Country', '1970–1975']].head().style.hide(axis=0) # 2005–2010 # podemos hacer que muestre cabecera y darle un estilo para que el axiis 0 no lo muestre
```

```python
# Índice de natalidad por país entre los años 1985 a 1990 ordenado de más a menos (primeros resultados)
fertility_rate[['Country', '1985–1990']].sort_values(by= '1985–1990', ascending= False ).head().style.hide(axis=0) 
```

```python
fertility_rate[['Country', '1985–1990']].sort_values(by= '1985–1990', ascending= False ).tail().style.hide(axis=0) # ahora por debajo
```

```python
# Al venir de HTML todos los datos son strings, si queremos números tenemos que transformarlos
# Vamos a transformar todas las columnas (desde la segunda, 0 es indice la 1 es el país) a valores numéricos
fertility_rate = fertility_rate[1:][:].apply(pd.to_numeric, errors='coerce')  # primera fila, coerce es para pasarlo a un NaN
```

```python
# Calcular la media del índice de natalidad para cada año
fertility_rate.mean()[1:] # a partir de la segunda fila
```

```python
# Mostramos un fráfico del resultado
import matplotlib.pyplot as plt

plt.style.use('ggplot')

fig, ax = plt.subplots(figsize= (16,9), dpi= 128)

ax.plot(fertility_rate.mean()[1:], color= 'blue', linewidth= 7)

ax.set_title('Natalidad Mundial / Periodos 1950 - 2015')
ax.set_xlabel('', fontsize= 14)
fig.autofmt_xdate()
ax.set_ylabel('Media natalidad mundial', fontsize= 14)
ax.tick_params(labelsize= 12)
plt.show()
```
