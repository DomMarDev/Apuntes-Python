## Ejemplo práctico: Caminatas aleatórias

Crear la calse Caminata_aleatoria()

- Nuestra clase tomara decisiones aleatorias sobre la dirección que debería llevar.
- Necesitamos tres atributos:
    - Una variable para realizar un seguimiento del número de puntos de la caminata
    - Dos listas para guardar los valores del eje de coordenadas x e y de cada punto del camino
- Dos métodos:
    -\_\_init\_\_()
    -trazado_paseo(), que va a calcular los puntos

```python
# Vamos a instalar una biblioteca de barra de progreso. la librería tqdm para visualizar procesos de creación (funciona con bucles)
!pip install tqdm
```

```python
# Importamos tres librerías
from random import choice
from tqdm import tqdm
import matplotlib.pyplot as plt

class Caminata_aleatoria:
    ''' Inicializamos los atributos de la caminata'''
    def __init__(self, num_puntos = 5000):
        self.num_puntos = num_puntos

        # iniciamos los ejes del camino (x e y)
        self.valores_x = [0] # punto inicial
        self.valores_y = [0] 

    def trazado_paseo(self):
        ''' Calculamos todos los puntos de la caminata'''

        # Pasos hasta que la caminata alcanza la longitud definida (5000)
        with tqdm(total = self.num_puntos) as pbar:
            while len(self.valores_x) < self.num_puntos:

                # decidimos en qué dirección vamos a ir y cuánto vamos a avanzar en esa dirección
                direccion_x = choice([1,-1]) # adelante o atrás
                distancia_x = choice([0,1,2,3,4]) # pasos que realiza
                pasos_x = direccion_x * distancia_x

                # decidimos en qué dirección vamos a ir y cuánto vamos a avanzar en esa dirección
                direccion_y = choice([1,-1]) # adelante o atrás
                distancia_y = choice([0,1,2,3,4]) # pasos que realiza
                pasos_y = direccion_y * distancia_y

                # rechazamos los pasos que no nos llevan a ningún lado
                if pasos_x == 0 and pasos_y == 0:
                    continue
                
                # calculamos la nueva posición
                x = self.valores_x[-1] + pasos_x # ultima posición del paso + el paso nuevo
                y = self.valores_y[-1] + pasos_y # ultima posición del paso + el paso nuevo

                self.valores_x.append(x) # haces append a los valores de x el valor nuevo
                self.valores_y.append(y)

                pbar.update(1) # la barra de progreso la llenamos en 1


```

```python
# probamos el funcionamiento del script anterior
micaminata = Caminata_aleatoria()
micaminata.trazado_paseo()
```

Trazamos la caminata

```python
plt.style.use('classic')
fig, ax = plt.subplots()

ax.scatter(micaminata.valores_x, micaminata.valores_y, s= 15 )
ax.set_aspect('equal')

plt.show()
```

### Múltiples caminatas (vamos a hacer un script)

```python
import matplotlib.pyplot as plt

while True:
    puntos = int(input('¿Cuantos puntos quiere en la caminata?'))

    otracaminata = Caminata_aleatoria(puntos) # pordefecto eran 5000
    otracaminata.trazado_paseo()

    plt.style.use('classic')

    fig, ax = plt.subplots()

    #Trazamos un mapa de color
    numero_puntos = range(otracaminata.num_puntos)
    ax.scatter(otracaminata.valores_x, otracaminata.valores_y, c= numero_puntos, cmap=plt.cm.inferno, edgecolors= 'none', s= 1)

    # marcamos el punto de inicio y el punto de final
    ax.scatter(0,0, color= 'red', edgecolors='None', s=100)
    ax.scatter(otracaminata.valores_x[-1], otracaminata.valores_y[-1], color= 'blue', edgecolors='None', s=100)

    # eliminamos los jes del trazado
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    ax.set_aspect('equal')

    plt.show()

    seguimos = input('¿Continuar? s/n').lower()
    if seguimos == 'n':
        break
```

### Creamos la clase Dado

```python
from random import randint

class Dado:
    ''' Una clase que representa un dado de x caras'''

    def __init__(self, num_caras= 6):
        ''' Por defecto el dado tiene 6 caras'''
        self.num_caras = num_caras

    def tirada(self):
        ''' Devuelve un número aleatorio entre 1 y el número de caras'''
        return randint(1, self.num_caras)
    
```

Comprobamos la clase haciendo una tirada de dado y devolviendo el resultado

```python
miDado = Dado() #instanciamos

misTiradas = [] # acumular datos del dado

for tirada in range(1000):
    tirada = miDado.tirada()
    misTiradas.append(tirada)

print(misTiradas[:100]) # que muestre las 100 primeras
```

### Analisis de los resultados

```python
frecuencias = []
analisis_resultado = range(1, miDado.num_caras+1) # son los 6 lados del dado
for valor in analisis_resultado:
    frecuencia = misTiradas.count(valor)
    frecuencias.append(frecuencia)
print(frecuencias)
```
