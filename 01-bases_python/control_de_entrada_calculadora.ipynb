```python
# Funciones

# Introducir datos

def introducir_datos():
    num1 = int(input('Dame el número A'))
    num2 = int(input('Dame el número B'))
    
    return num1, num2
    
#Son numeros?

def son_numeros(num1, num2):
    control = True

    if num1.isdigit() and num2.isdigit():
        control = True
    else:
        control = False
    
    return control

     




#Suma

def suma_valores(num1, num2):

    return print(num1 + num2)
  

#Resta

def resta_valores(num1, num2):

    return print(num1 - num2)




#Multiplicación
def multi_valores(num1, num2):

    return print(num1 * num2)

#División
def divi_valores(num1, num2):

    if num2 == 0:
        print('No se puede dividir por 0')
    else:
        return print(num1 / num2)


    

        



#Función principal

while True:
    prompt = '¿Qué quieres hacer? 1) Suma  2) Resta 3) Multiplicación 4) División 5) Salir'
    
    print(prompt)
    
    opcion = input('Entra la opción')

    match opcion:
        case '1':
           introducir_datos()
           suma_valores(num1,num2)
        case '2':
           introducir_datos()
           resta_valores(num1, num2)
        
        case '3':
           introducir_datos()
           multi_valores(num1, num2)
        case '4':
           introducir_datos()
           divi_valores(num1, num2)      
        case '5':
            print('Adiós Lucas')
            break
        case _:
            print('Opción no contemplada')
```

```python
# Calculadora

import os

def clear_screen():
    os.system('cis')

def pause_screen():
    print('Pulse enter para continuar')
    input()

def evaluar():
    while True:#Control de excepciones hacer una operación y si se da pasar de el
        try:
            numero1= float(input('Introduce el primer valor'))
            break
        except ValueError: #El valor del error es ValueError
            print('Introduce un valor numérico')
            pause_screen()
    while True:
        try:
            numero2= float(input('Introduce el segundo valor'))
            break
        except ValueError:
            print('Introduce un valor numérico')
            pause_screen()         
    return numero1, numero2

def sumar():
    numeros = evaluar()
    resultado = sum(numeros)
    return resultado

def restar():
    numeros = evaluar()
    resultado = numeros[0] - numeros[1]
    return resultado    

def multiplicar():
    numeros = evaluar()
    resultado = numeros[0] * numeros[1]
    return resultado 

def dividir():
    

    while True:
        numeros = evaluar()
        try:
            resultado = numeros[0] / numeros[1]
            return resultado
        except ZeroDivisionError:
            print('No se puede dividir por cero, introduce otro número')


# Programa Principal

while True:
    print(''' 
1) Suma  
2) Resta 
3) Multiplicación 
4) División 
          
5) Salir
''' )
    opcion = input('Dame una opcion a elegir')

    match(opcion):
        case '1':
            print(' Has elegido la suma')
            print(f' El resultado de la suma es: {sumar()}')

            pause_screen()
            clear_screen()

        case '2':
            print(' Has elegido la resta')
            print(f' El resultado de la resta es: {restar()}')
            pause_screen()
            clear_screen()
        case '3':
            print(' Has elegido la multiplicación')
            print(f' El resultado de la multiplicación es: {multiplicar()}')
            pause_screen()
            clear_screen()
        case '4':
            print(' Has elegido la división')
            print(f' El resultado de la división es: {dividir()}')
            pause_screen()
            clear_screen()
        case '5':
            print('adios')
            break
        case _:
            print('opción no contemplada')



```
